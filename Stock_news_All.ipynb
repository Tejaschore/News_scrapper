```python
def scrape_news(urls,tag):

  try:
    response = requests.get(urls)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    # Extract news content based on the given tag
    content = soup.find_all(tag)
    news = ' '.join([item.get_text(strip=True) for item in content])
    return news if news else "No news found"
  except Exception as e:
        return f"Error: {e}"
```

```python
news_df=websites.copy()
news_df
```

```python
news_df['news'] = news_df.apply(lambda row: scrape_news(row['weblink'], row['tag']), axis=1)
news_df
```
